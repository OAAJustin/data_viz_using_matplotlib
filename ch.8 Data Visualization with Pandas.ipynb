```python
# Import Libraries
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

### The DataFrame.plot() Function

- The plot() function accepts x and y features, and a kind argument. To avoid the kind argument, you casn also call DataFrame.plot.kind() instead, and pass in the x and y features.
- The accepted values for the kind argument are: line, bar, barh (horizontal bar), hist, box, kde, density (synonym for kde), area, pie, scatter and hexbin (similar to a Heatmap).

```python
# Reference the dataset file
file = 'gdp_csv.csv'

# Create the dataframe
df = pd.read_csv(file)

# EU dataframe
df_eu = df.loc[df['Country Name'] == 'European Union']

# Bar plot for EU dataframe
df_eu.plot.bar(x = 'Year', y = 'Value')

# Line plot for EU dataframe
df_eu.plot.line(x = 'Year', y = 'Value')

# Box plot for EU dataframe
df_eu.plot.box(x = 'Value')

# Show the figure
plt.show()
```

```python
# Plot on one figure
# Reference the dataset file
file = 'gdp_csv.csv'

# Create the dataframe
df = pd.read_csv(file)

# EU dataframe
df_eu = df.loc[df['Country Name'] == 'European Union']

# Create the figure and axes objects
fig, ax = plt.subplots(3, 1)

# Box plot for EU dataframe
df_eu.plot.box(x = 'Value', ax = ax[0])
# Line plot for EU dataframe
df_eu.plot.line(x = 'Year', y = 'Value', ax = ax[1])
# Bar plot for EU dataframe
df_eu.plot.bar(x = 'Year', y = 'Value', rot = 45, ax = ax[2])

# Show the figure
plt.show()
```

### Pandas' Plotting Module

### Bootstrap Plot

- Bootstrapping is the process of randomly sampling (with replacement) a dataset, and calculating measures of accuracy such as bias, variance and confidence intervals for the random samples.
- 'With replacement', means that each randomly selected element can be selected again. 'Without replacement' means that after each randomly selected element, it is removed from the pool for the next sample.
- A Bootstrap Plot bootstraps the mean, median and mid-range statistics of a dataset, based on the sample size, shown via plt.show()
- The default arguments for size and samples are 50 and 500 respectively.

```python
# Reference the dataset file
file = 'gdp_csv.csv'

# Create the dataframe
df = pd.read_csv(file)

# EU dataframe
df_eu = df.loc[df['Country Name'] == 'European Union']

# Bootstrap plot for EU dataframe of values column
pd.plotting.bootstrap_plot(df_eu['Value'])

# Show the figure
plt.show()
```

### Autocorrelation Plot

 - Autocorrelation Plots are used to check for data randomness, for time-series data. Multiple autocorrelations are calculated for differing timestamps, and if the data is truly random - the correlation will be near zero. If not - the correlation will be larger than zero.
 - The Autocorrelation Plot for the random_series should revolve around 0, since it's random data, while the plot for the Value feature won't. 
 

```python
# Reference the dataset file
file = 'gdp_csv.csv'

# Create the dataframe
df = pd.read_csv(file)

# EU dataframe
df_eu = df.loc[df['Country Name'] == 'European Union']

# Create a random series of numbers between the range of 100 
# and the size limit of 50
random_series = pd.Series(np.random.randint(0, 100, size = 50))

# Autocorrelation plot for EU dataframe of values dataframe
pd.plotting.autocorrelation_plot(df_eu['Value'])
# Autocorrelation plot for random series
pd.plotting.autocorrelation_plot(random_series)

# Show the figure
plt.show()
```

### Scatter Matrices

 - Scatter Matrices plot a grid of Scatter Plots for all features against all features. Scatter Matrices are also known as Pair Plots. Seaborn offers a pairplot()

```python
# Reference the dataset file
file = '/Users/Justin/dev/data_viz_in_matplotlib/data_viz_using_matplotlib/worldhappiness/2019.csv'

# Create the dataframe
df = pd.read_csv(file)

# Scatter matrix
axes = pd.plotting.scatter_matrix(df, diagonal = 'hist')

# Set rotation and set label position
for ax in axes.flatten():
    ax.yaxis.label.set_rotation(0)
    ax.yaxis.label.set_ha('right')

# Show the figure
plt.show()
```

- You can pass in the diagonal argument, which accepts 'hist' or 'kde' to specify what type of distribution plot you'd like to plot on the diagonal, as well as alpha, specifying the translucency of the markers in the Scatter Plots
